```python
import pandas as pd
from pathlib import Path

data_path = Path("../data/transactions.csv")
```

```python
df = pd.read_csv(data_path)

# Add transaction_id if it does not exist
if "transaction_id" not in df.columns:
    df.insert(0, "transaction_id", range(1, len(df) + 1))

df
```

```python
new_transaction = {
    "transaction_id": df["transaction_id"].max() + 1,
    "date": "2025-03-01",
    "category":"Food",
    "amount": 25.50
}

df = pd.concat([df, pd.DataFrame([new_transaction])], ignore_index=True)
df.to_csv(data_path, index=False)

df.tail()
```

```python
transaction_id = 1
new_amount = 45.00

if transaction_id in df["transaction_id"].values:
    df.loc[df["transaction_id"] == transaction_id, "amount"] = new_amount
else:
    print("Transaction ID not found")
df.loc[df["transaction_id"] == transaction_id, "amount"] = new_amount

df.to_csv(data_path, index=False)

df.head()
```

```python
transaction_id = 2

df = df[df["transaction_id"] != transaction_id]

df = df.reset_index(drop=True)

df.to_csv(data_path, index=False)

df
```
