# Transaction Data Setup
This cell imports the required libraries and defines the path to the transaction dataset. 
Basic error handling ensures the file path is valid before loading the data later.

```python
import pandas as pd
from pathlib import Path
from IPython.display import display

try:
    data_path = Path("../data/transactions.csv")
except FileNotFoundError:
    print("Error: The data file was not found. Please check the path.")
except KeyError as e:
    print(f"Missing column in dataset: {e}")
except Exception as e:
    print(f"Unexpected error occurred: {e}")
```

# Load Dataset
This cell loads the CSV file into a pandas DataFrame and performs initial validation. 
If the dataset is empty or cannot be loaded, an error message is displayed. 

```python
try:
    df = pd.read_csv(data_path)
    if df.empty:
        print("No data available")    
    display(df)  # explicitly renders the DataFrame
except FileNotFoundError:
    print("Error: The data file was not found. Please check the path.")
except KeyError as e:
    print(f"Missing column in dataset: {e}")
except Exception as e:
    print(f"Unexpected error occurred: {e}")
```

# Add a New Transaction
This cell demonstrates how to append a new transaction to the dataset. 
Error handling ensures the DataFrame exists and that the required fields are valid before insertion.

```python
try:
    new_transaction = {
        "date": "2025-03-01",
        "category":"Food",
        "amount": 25.50
    }
    
    df = pd.concat([df, pd.DataFrame([new_transaction])], ignore_index=True)
    df.to_csv(data_path, index=False)
    
    display(df.tail())
except FileNotFoundError:
    print("Error: The data file was not found. Please check the path.")
except KeyError as e:
    print(f"Missing column in dataset: {e}")
except Exception as e:
    print(f"Unexpected error occurred: {e}")
```

# Update a Transaction
This cell updates the amount of an existing transaction. 
If the ID does not exist, the code safely handles the error and informs the user.

```python
try:
    excel_row = 3612  # Excel row number (starts from 2, since row 1 is header)
    new_amount = 45.00
    df_index = excel_row - 2  # Convert to DataFrame index (0-based)
    
    if df_index in df.index:
        df.loc[df_index, "amount"] = new_amount
    else:
        print("Row number not found")
    
    df.to_csv(data_path, index=False)
    
    display(df)
except FileNotFoundError:
    print("Error: The data file was not found. Please check the path.")
except KeyError as e:
    print(f"Missing column in dataset: {e}")
except Exception as e:
    print(f"Unexpected error occurred: {e}")
```
